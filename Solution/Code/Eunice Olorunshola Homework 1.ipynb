```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os
from PIL import Image
#enable plots in the notebook
%matplotlib inline 

#load the image
fname = 'image.jpg'
img = Image.open(fname)
# display the image on the console
plt.imshow(img)


#convert the image data into a numpy matrix 
imgmat = np.array(list(img.getdata(band = 0)), float)
#Reshape according to original image dimensions
imgmat.shape = (img.size[1], img.size[0])
imgmat = np.matrix(imgmat)
plt.figure()
#Display the original image in gray
plt.imshow(imgmat, cmap = plt.get_cmap('gray'))
plt.title("Image showing Grayscale after converted")


#Single value decomposition 
U, S, Vt = np.linalg.svd(imgmat)


#Here are the rank-one matrix from the Principal componenet approximation ,it shows the reconstructed images
imgrecon = np.matrix(U[:, :1]) * np.diag(S[:1]) * np.matrix(Vt[:1,:])
plt.figure()
#Display the reconstructed images from the k singluar values in gray
plt.imshow(imgrecon, cmap = plt.get_cmap('gray'))
plt.title("k=1")
plt.show()
        
imgrecon = np.matrix(U[:, :5]) * np.diag(S[:5]) * np.matrix(Vt[:5,:])
plt.figure()
plt.imshow(imgrecon, cmap = plt.get_cmap('gray'))
plt.title("k=5")
plt.show()
        
imgrecon = np.matrix(U[:, :10]) * np.diag(S[:10]) * np.matrix(Vt[:10,:])
plt.figure()
plt.imshow(imgrecon, cmap = plt.get_cmap('gray'))
plt.title("k=10")
plt.show()
        
imgrecon = np.matrix(U[:, :20]) * np.diag(S[:20]) * np.matrix(Vt[:20,:])
plt.figure()
plt.imshow(imgrecon, cmap = plt.get_cmap('gray'))
plt.title("k=20")
plt.show()
        
        
imgrecon = np.matrix(U[:, :40]) * np.diag(S[:40]) * np.matrix(Vt[:40,:])
plt.figure()
plt.imshow(imgrecon, cmap = plt.get_cmap('gray'))
plt.title("k=40")
plt.show()
        
```
